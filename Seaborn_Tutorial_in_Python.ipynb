***Seaborn : Data Visualization tool***

```python
# Importing the libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

***Note : Seaborn has some built-in datasets***

```python
# total bill vs tip dataset
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# setting a theme for the plots
sns.set_theme()
```

```python
# visualize the data
sns.relplot(data=tips, x ='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',fit_reg=False,markers=['^','s'],palette='ocean',row='sex',col='smoker')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=False)
plt.grid()
plt.axhline(20,color='black')
plt.show()
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='sex')
plt.grid()
plt.axhline(20,color='black')
plt.show()
```

```python
# HEAT MAP
sns.heatmap(tips.corr(),annot=True, cmap = 'Blues')
plt.figure(figsize=(5,5))
plt.show()
```

```python
sns.countplot(x='day',data=tips,hue='sex')
plt.show()
```

```python
sns.countplot(x='sex',data=tips,hue='day')
plt.show()
```

```python
sns.countplot(x='size',data=tips,hue='sex')
plt.show()
```

```python
sns.pairplot(data=tips,hue='sex')
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
sns.violinplot(y='day',x='total_bill',data=tips)
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,whis=False)
plt.show()
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker',split=True)
plt.show()
```

```python
a=tips.groupby('sex').mean()['total_bill']
```

```python
a.plot(kind='bar')
plt.show()
```

```python
tips.groupby('day').mean()['total_bill'].plot(kind='bar')
plt.show()
```
