# **Scatter Plot**

***Used when relationship between 2 sets of values to be compared and to see if they are corelated.***

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.use('seaborn')
```

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 6, 4, 3, 2, 9, 9, 7]
y = [7, 5, 4, 2, 1, 6, 8, 9, 3, 4, 4, 2, 2, 5, 1, 6, 8, 6, 6, 4, 4, 3, 2]
```

```python
plt.scatter(x,y)
```

***Changing the size of dots***

```python
plt.scatter(x,y, s=100)
```

***Changing color***

```python
plt.scatter(x,y, s=100, c='green')
```

***Instead of dots we can use other markers like X or o***

```python
plt.scatter(x,y, s=100, c='green', marker='x')
```

***Giving Edgecolor***

```python
plt.scatter(x,y, s=100, c='green', edgecolor='black', linewidth=1)
```

***Adding new features***

```python
plt.scatter(x,y, s=100, c='green', edgecolor='black', linewidth=1, alpha=0.75)
```

***Adding additional information to dataset using colors***


```python
colors = [7,5,9,7,5,7,2,5,3,7,1,2,8,1,9,2,5,6,7,5,3,1,7]
plt.scatter(x,y, s=100, c=colors, edgecolor='black', linewidth=1)
```

***Changing the shades of grey***

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 6, 4, 3, 2, 9, 9, 7]
y = [7, 5, 4, 2, 1, 6, 8, 9, 3, 4, 4, 2, 2, 5, 1, 6, 8, 6, 6, 4, 4, 3, 2]

colors = [7,5,9,7,5,7,2,5,3,7,1,2,8,1,9,2,5,6,7,5,3,1,7]
plt.scatter(x,y, s=100, c=colors, cmap='Greens', edgecolor='black', linewidth=1)

cbar = plt.colorbar()
cbar.set_label('Rating')
```

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 6, 4, 3, 2, 9, 9, 7]
y = [7, 5, 4, 2, 1, 6, 8, 9, 3, 4, 4, 2, 2, 5, 1, 6, 8, 6, 6, 4, 4, 3, 2]

colors = [7,5,9,7,5,7,2,5,3,7,1,2,8,1,9,2,5,6,7,5,3,1,7]
plt.scatter(x,y, s=100, c=colors, cmap='rainbow', edgecolor='black', linewidth=1)

cbar = plt.colorbar()
cbar.set_label('Rating')
```

***Let's work on real data***

```python
import pandas as pd
import io
```

```python
from google.colab import files
  
  
uploaded = files.upload()
```

```python
data = pd.read_csv(io.BytesIO(uploaded['Scatter Plot Data.csv']))
```

```python
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']
```

```python
plt.scatter(view_count, likes, s=100, edgecolor='black', linewidth=1, alpha=0.75)
```

***Not looking a good plot because of outliers. Hence lets scale the plot to log scale.***

```python
plt.scatter(view_count, likes, s=100, edgecolor='black', linewidth=1, alpha=0.75)

plt.xscale('log')
plt.yscale('log')
```

***Lets use ratio of likes/dislikes to represent more information in plot using colors.***

***We can more information about matplotlib from : https://matplotlib.org/stable/tutorials/colors/colormaps.html***

```python
plt.scatter(view_count, likes, s=100, c=ratio, cmap = 'Spectral', edgecolor='black', linewidth=1, alpha=0.75)

cbar = plt.colorbar()
cbar.set_label('Likes/Dislikes Ratio')

plt.xscale('log')
plt.yscale('log')
```
